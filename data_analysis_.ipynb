```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the Excel file into a DataFrame
excel_file_path = '/content/drive/MyDrive/data of all years .xlsx'
df = pd.read_excel(excel_file_path)

# Set seaborn style
sns.set(style="whitegrid")

# Create subplots for each column
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Plot Age distribution
sns.histplot(df['Age'], bins=20, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Age Distribution')

# Plot Sex distribution
sns.countplot(x='Sex', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Sex Distribution')

# Plot Diagnosis distribution with improved label rotation and spacing
diagnosis_plot = sns.countplot(x='Diagnosis', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Diagnosis Distribution')
diagnosis_plot.set_xticklabels(diagnosis_plot.get_xticklabels(), rotation=45, ha='right', fontsize=8)  # Rotate x-axis labels for better visibility

# Additional plot (you can customize this based on your data)
# For example, let's plot a scatter plot of Age vs Diagnosis
sns.scatterplot(x='Age', y='Diagnosis', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Scatter Plot: Age vs Diagnosis')

# Adjust layout
plt.tight_layout()

# Save the plots to a JPG file
#output_file_path = '/content/drive/MyDrive/plots.jpg'
#plt.savefig(output_file_path, format='jpg')

# Show the plots (optional)
plt.show()
```

```python
import pandas as pd

# Provide the path to your CSV file
csv_file_path = "/content/drive/MyDrive/data of all years .xlsx"

# Read the CSV file into a DataFrame
df = pd.read_excel(csv_file_path)

# List of elements in the middle region
middle_region_elements = [
    "Misurata", "Misilata", "Sert", "Ben Walid", "Khums", "Zliten", "Emsehel", "Kaser Kayar","Gamata","Abugrain",
    "Kasar Kayar", "Garaboli", "Tarhona", "Al Amamra", "Al Alloos", "Kaam(Zliten)", "Misallata","Garaboli","Ganema","Tawargha","Al Allous","Zintan","Zultan"
]

# Filter the DataFrame to include only the middle region elements
middle_region_df = df[df['Address'].isin(middle_region_elements)]

# Count how many times each diagnosis appears in the middle region
diagnosis_counts = middle_region_df['Diagnosis'].value_counts()

# Calculate the percentages
diagnosis_percentages = (diagnosis_counts / diagnosis_counts.sum()) * 100

# Create a DataFrame to display counts and percentages
diagnosis_summary = pd.DataFrame({'Diagnosis Counts': diagnosis_counts, 'Percentage': diagnosis_percentages})

# Print the diagnosis counts and percentages
print("Diagnosis counts and percentages in the middle region:")
print(diagnosis_summary)
```
